### Read data from 2 .csv

```python
import numpy as np

# Read the CSV file, skipping the first 4 rows (header)
data = np.loadtxt("data/001MoDe_R1.csv", delimiter=",", skiprows=4)
data = data[:-1, :]

# Extract columns
timestamp = data[:, 0]
x = data[:, 1]
y = data[:, 2]
in_target = data[:, 3]

marker = "data/001MoDe_R1.marker.csv"
with open(marker, "r") as f:
    lines = f.readlines() # Read all lines from the marker file

# Print the shape of the data array to verify loading
print(data.shape)

# Display the first 20 lines to understand the structure
for l in lines[:20]:
    print(l.strip())
```

### Start and End period of record

```python
# On lit le fichier principal
timestamp = data[:, 0]
time = (timestamp - timestamp[0]) / 1000.0  # en secondes, depuis le début

# On lit les marqueurs
record_start = []
record_end = []

for line in lines:
    parts = line.strip().split(",")
    if len(parts) < 3:
        continue
    try:
        ts = int(parts[1])  # timestamp du marker en ms
    except ValueError:
        continue

    if "DoCycleChange:DoRecord" in line:
        record_start.append(ts)
    elif "DoCycleChange:DoPause" in line:
        record_end.append(ts)

# On suppose que les timestamps des markers sont dans le même référentiel temporel que ceux du CSV principal
# => On les convertit en secondes par rapport au même zéro (timestamp[0])

start_record = [(t - timestamp[0]) / 1000.0 for t in record_start]
stop_record  = [(t - timestamp[0]) / 1000.0 for t in record_end]

print("Start record times (s):", start_record)
print("End record times (s):", stop_record)
```
