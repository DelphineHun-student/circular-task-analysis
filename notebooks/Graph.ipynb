```python
import numpy as np

# Read the CSV file, skipping the first 4 rows (header)
data = np.loadtxt("data/001MoDe_R1.csv", delimiter=",", skiprows=4)

# Extract columns
timestamp = data[:, 0]
x = data[:, 1]
y = data[:, 2]
in_target = data[:, 3]

file_path = "data/001MoDe_R1.marker.csv"
with open(file_path, "r") as f:
    lines = f.readlines()

# Print the shape of the data array to verify loading
print(data.shape)

# Display the first 20 lines to understand the structure
for l in lines[:20]:
    print(l.strip())
```

```python
record_start = []
record_end = []

for line in lines:
    parts = line.strip().split(",")
    if len(parts) < 3:
        continue 
    try:
        timestamp = int(parts[1])
    except ValueError:
        continue 

    if "DoCycleChange:DoRecord" in line:
        record_start.append(timestamp)
    elif "DoCycleChange:DoPause" in line:
        record_end.append(timestamp)

# Normalize times to start from zero and convert to seconds
if len(record_start) > 0:
    t0 = min(record_start + record_end)
    start_record = [(t - t0) / 1000.0 for t in record_start]
    stop_record = [(t - t0) / 1000.0 for t in record_end]
else:
    raise ValueError("Aucun événement 'DoRecord' détecté dans le fichier.")

# Print the extracted times
print("Start record times (s):", start_record)
print("End record times (s):", stop_record)
```
